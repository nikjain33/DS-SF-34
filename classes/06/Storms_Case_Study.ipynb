```python
df = df.set_index('REFNUM')
```

```python
df = df.drop(['COUNTYENDN' , 'COUNTY_END'], axis = 1)
```

```python
df = df[ ['EVTYPE', 'FATALITIES', 'INJURIES', 'PROPDMG', 'PROPDMGEXP', 'CROPDMG', 'CROPDMGEXP'] ]
df
```

```python
df_1 = df[df.PROPDMGEXP == 'K' or 'k']
df_1 = df_1.drop('PROPDMGEXP', axis = 1)

# Scaling property damage to K
df_1.PROPDMG *= 1000

df_2 = df[df.PROPDMGEXP == 'M' or 'm']
df_2 = df_2.drop('PROPDMGEXP', axis = 1)

#Scaling property damage to M
df_2.PROPDMG *= 10 ** 6

df_3 = df[(df.PROPDMGEXP == 'B') | (df.PROPDMGEXP == 'b')]
df_3 = df_3.drop('PROPDMGEXP', axis = 1)

#Scaling property damage to B
df_3.PROPDMG *= 10 ** 9

df_4 = df[df.PROPDMGEXP < 9]
df_4 = df_4.drop('PROPDMGEXP', axis = 1)

#Scaling property damage to 10
df_4.PROPDMG *= 10

df_5 = df[df.PROPDMGEXP == 'H' or 'h']
df_5 = df_5.drop('PROPDMGEXP', axis = 1)

#Scaling property damage to 100
df_5.PROPDMG *= 100
```
