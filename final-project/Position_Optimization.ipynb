```python
def OPPPtest(player):
    y1 = cs.OPPP[cs[player] == 1].values
    y2 = cs.OPPP[cs[player] == 0].values
    t_val, p_val = ttest_ind(y1, y2)
    print 't_test results:'
    print 't-value', t_val
    print 'p-value', p_val
```
